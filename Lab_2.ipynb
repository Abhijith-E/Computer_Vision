```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
video_path = "ConjuringSampleVideo.mp4"
cap = cv2.VideoCapture(video_path)
```

```python
output_folder = "frames"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
frame_count = 0 

while True:
    ret, frame = cap.read()  # Read a frame
    if not ret:  # Stop when video ends
        break

    frame_filename = f"{output_folder}/frame_{frame_count}.jpg"
    cv2.imwrite(frame_filename, frame)  # Save the frame

    print(f"Saved: {frame_filename}")  # Print confirmation
    frame_count += 1  # Increment counter
```
