```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('HappyFish.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
identity_kernel = np.array([[0, 0, 0],
                            [0, 1, 0],
                            [0, 0, 0]])
```

```python
edge_detection_kernel = np.array([[-1, -1, -1],
                                  [-1,  8, -1],
                                  [-1, -1, -1]])
```

```python
sharpening_kernel = np.array([[0, -1, 0],
                              [-1, 5, -1],
                              [0, -1, 0]])
```

```python
box_blur_kernel = np.ones((3, 3), dtype=np.float32) / 9.0
```

```python
gaussian_blur_kernel = np.array([[1, 2, 1],
                                 [2, 4, 2],
                                 [1, 2, 1]], dtype=np.float32) / 16.0
```

```python
def apply_operation(image, kernel, operation='convolution'):
    kernel = np.flipud(np.fliplr(kernel)) if operation == 'convolution' else kernel
    output = cv2.filter2D(image, -1, kernel)
    return output
```

```python
identity_result = apply_operation(image, identity_kernel, 'convolution')
edge_detection_result = apply_operation(image, edge_detection_kernel, 'convolution')
sharpening_result = apply_operation(image, sharpening_kernel, 'convolution')
box_blur_result = apply_operation(image, box_blur_kernel, 'convolution')
gaussian_blur_result = apply_operation(image, gaussian_blur_kernel, 'convolution')
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
```

```python
plt.subplot(2, 3, 2)
plt.title('Identity Kernel')
plt.imshow(identity_result, cmap='gray')
```

```python
plt.subplot(2, 3, 3)
plt.title('Edge Detection Kernel')
plt.imshow(edge_detection_result, cmap='gray')
```

```python
plt.subplot(2, 3, 4)
plt.title('Sharpening Kernel')
plt.imshow(sharpening_result, cmap='gray')
```

```python
plt.subplot(2, 3, 5)
plt.title('Box Blur Kernel')
plt.imshow(box_blur_result, cmap='gray')
```

```python
plt.subplot(2, 3, 6)
plt.title('Gaussian Blur Kernel')
plt.imshow(gaussian_blur_result, cmap='gray')
```

```python
plt.tight_layout()
plt.show()
```
