```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input grayscale image
img = cv2.imread('logo.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded properly
if img is None:
    print("Error: Image not found!")
    exit()

# Q1: Gaussian filter with different sigma and border styles
sigma_values = [1, 3, 5]
border_styles = {
    'BORDER_REPLICATE': cv2.BORDER_REPLICATE,
    'BORDER_REFLECT': cv2.BORDER_REFLECT,
    'BORDER_CONSTANT': cv2.BORDER_CONSTANT
}

# Loop through each sigma and border type
for sigma in sigma_values:
    fig, axes = plt.subplots(1, len(border_styles) + 1, figsize=(15, 5))
    
    # Show the original image
    axes[0].imshow(img, cmap='gray', vmin=0, vmax=255)
    axes[0].set_title("Original Image")
    axes[0].axis("off")
    
    for i, (bname, btype) in enumerate(border_styles.items()):
        blurred = cv2.GaussianBlur(img, (0, 0), sigmaX=sigma, sigmaY=sigma, borderType=btype)
        
        # Show the blurred image
        axes[i + 1].imshow(blurred, cmap='gray', vmin=0, vmax=255)
        axes[i + 1].set_title(f'Sigma {sigma}\n{bname}')
        axes[i + 1].axis("off")
    
    # Show the figure with all images
    plt.show()
```
