```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image_path = 'boat.png'  # Replace with your image path
image = Image.open(image_path)
```

```python
# Define the new size (e.g., half the original size)
new_size = (image.width // 2, image.height // 2)
```

```python
# Resize using different interpolation techniques
# i) Bilinear Interpolation
bilinear_image = image.resize(new_size, Image.BILINEAR)
```

```python
# ii) Bicubic Interpolation
bicubic_image = image.resize(new_size, Image.BICUBIC)
```

```python
# iii) Nearest Neighbor Interpolation
nearest_image = image.resize(new_size, Image.NEAREST)
```

```python
# iv) Lanczos Interpolation
lanczos_image = image.resize(new_size, Image.LANCZOS)
```

```python
# Display the images
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.imshow(bilinear_image)
plt.title('Bilinear Interpolation')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(bicubic_image)
plt.title('Bicubic Interpolation')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(nearest_image)
plt.title('Nearest Neighbor Interpolation')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(lanczos_image)
plt.title('Lanczos Interpolation')
plt.axis('off')

plt.tight_layout()
plt.show()
```
